```python
Given the head of a singly linked list, return true if it is a palindrome.


Example 1:
Input: head = [1,2,2,1]
Output: true

Example 2:
Input: head = [1,2]
Output: false
 

Constraints:
The number of nodes in the list is in the range [1, 105].
0 <= Node.val <= 9
 
Follow up: Could you do it in O(n) time and O(1) space?
```

```python
This is a cool trick but this implementation has O(N) space complexity

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        l=[]
        
        if head==None and head.next==None:
            return True
        
        while head:
            l.append(head.val)
            head=head.next
            
        beg=0
        end=len(l)-1
        while beg<end:
            if l[beg]!=l[end]:
                return False
            
            beg+=1
            end-=1
        return True  
            
```

```python
#So let's try a solution with O(1) space complexity

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        
        #first we are going to find the mid point
        slow=head
        fast=head
        while fast!=None and fast.next!=None:
            slow=slow.next
            fast=fast.next.next
            
        if fast!=None:#which means the linkedlist is odd numbered and so we have to push the mid point slow to next pointer
            slow=slow.next
            
        #template function to reverse the linked-list from the midpoint
        def reverse_fn(head):
            prev=None
            while head:
                next_node=head.next
                head.next=prev
                prev=head
                head=next_node
            return prev
        
        head2=reverse_fn(slow)#we are actually reversing the linked-list from the midpoint 'slow' here. 'slow' is our midpoint
        while head2!=None:
            if head.val!=head2.val:
                return False
            head=head.next
            head2=head2.next
        return True
            
Time complexity: O(N)
Space complexity: O(1)
```
